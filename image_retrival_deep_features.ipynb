```python
test = graphlab.SFrame('image_test_data/')
train = graphlab.SFrame('image_train_data/')
```

```python
knn_model = graphlab.nearest_neighbors.create(train,
                                             features=['deep_features'],
                                             label='id')
```

```python
def get_image_id(query_results):
    return train.filter_by(query_results['reference_label'],'id')
```

```python
get_image_id(knn_model.query(cat))['image'].show()
```

```python
def show_neighbours(n):
    x = train[n:n+1]
    query_results = knn_model.query(x)
    return train.filter_by(query_results['reference_label'],'id')['image'].show()
```
